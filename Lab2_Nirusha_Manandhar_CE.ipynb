<b> Evaluation - holding out data and cross-validation </b>

What we really want to do is estimate the ability of the model to generalize to new instances. We should measure its performance in instances that the algorithm has not seen at training. Therefore, we are going to do a more rigorous evaluation and use held-out data. For this data is grouped into two groups; training data and testing data.

The accuracy on the training data, the training accuracy us almost always an overly optimistic estimate.

One possible problem : training accuracy could decrease since only half of data are used for training. And if further more data are used in training, low number of data are used in testing. Hence, the error estimation is performed on a very small number of examples. Ideally, all of the data used in training and testing is better, which is not possible. We can achieve a good approimation of this impossible ideal by a method called <b> cross-validation </b>

One simple form of cross-validation is <b> leave-one-out cross validation: </b>

```python
correct=0.0
```

```python
def fit_model(features, labels):
    '''Learn a simple threshold model'''
    best_acc = -1.0
    # Loop over all the features:
    for fi in range(features.shape[1]):
        thresh = features[:, fi].copy()
        # test all feature values in order:
        thresh.sort()
        for t in thresh:
            pred = (features[:, fi] > t)

            # Measure the accuracy of this 
            acc = (pred == labels).mean()

            rev_acc = (pred == ~labels).mean()
            if rev_acc > acc:
                acc = rev_acc
                reverse = True
            else:
                reverse = False
            if acc > best_acc:
                best_acc = acc
                best_fi = fi
                best_t = t
                best_reverse = reverse

    # A model is a threshold and an index
    return best_t, best_fi, best_reverse
    
```

```python
# This function was called ``apply_model`` in the first edition
def predict(model, features):
    '''Apply a learned model'''
    # A model is a pair as returned by fit_model
    t, fi, reverse = model
    if reverse:
        return features[:, fi] <= t
    else:
        return features[:, fi] > t
```

```python
def accuracy(features, labels, model):
    '''Compute the accuracy of the model'''
    preds = predict(model, features)
    return np.mean(preds == labels)
```

When using cross-validation, there is no circularity problem becuase each example was tested on a model which was built without taking hat datapoint into account. Therefore, the cross-validated estimate is a relaibel estimate of how well the models would generalize to new data.

The major problem with leave-one-out cross-validation is that we are now forced to erform many times more work. In fact, a whole new model should be learnt for each and every example and this will cost even more when data grows.

The training error can however bemisleadin, very-optimistic estimate of how well the model does. We must, instead, evaluate it on testing data that has not been used for training. In order to not waste too many examples on testing, a cross-validation schedule can get us the best of both training and testing.
